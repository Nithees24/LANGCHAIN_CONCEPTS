```python
import pdfplumber
import pandas as pd

with pdfplumber.open("input.pdf") as pdf:
  page = pdf.pages[6]

  table = page.extract_tables()[1]

  df = pd.DataFrame(table[1:], columns=table[0])

  print(df)
```
